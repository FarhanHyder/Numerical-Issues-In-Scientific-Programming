```python
from scipy.interpolate import lagrange
X = [300,400,500,600,700,800,900,1000,1100,1200,1300,1400,1500,1600,1700,1800,1900,2000]
Y = [0.024,0.035,0.046,0.058,0.067,0.083,0.097,0.111,0.125,0.140,0.155,0.170,0.186,0.202,0.219,0.235,0.252,0.269]

p = lagrange(X,Y)
Y_lag = []
for i in X:
    Y_lag.append(p(i))
plt.plot(X,Y_lag)
plt.show()
```
